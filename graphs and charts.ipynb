# Experiment 5

##### Aim : Plotting different types of graphs and charts by using matplotlib

```python
import matplotlib.pyplot as plt
```

```python
cities = ['Sangli', 'Satara', 'Kolhapur', 'Solapur', 'Karad', 'Pune', 'Mumbai', 'Kashamir', 'Tamilnadu', 'Hydrabad']
temperatures = [25, 30, 22, 28, 45, 12, 32, 78, 16, 25]
```

### Line Plot

```python
plt.figure(figsize=(8, 4))
plt.plot(cities, temperatures, marker='o',color='red', linestyle='-')
plt.grid(True)
plt.title('City Temperatures - Line Plot')
plt.xlabel('Cities')
plt.ylabel('Temperature (°C)')
plt.show()
```

### Bar Plot

```python
plt.figure(figsize=(8, 4))
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'lightcoral', 'lightblue', 'gold', 'lightpink', 'lightskyblue', 'plum']
plt.bar(cities, temperatures, color=colors )
plt.title('City Temperatures - Bar Plot')
plt.xlabel('Cities')
plt.ylabel('Temperature (°C)')
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(8, 4))
plt.scatter(cities, temperatures, color='maroon', marker='o')
plt.title('City Temperatures - Scatter Plot')
plt.xlabel('Cities')
plt.ylabel('Temperature (°C)')
plt.show()
```

### Pie Chart

```python
plt.figure(figsize=(6, 6))
colors = ['#87CEEB', '#7FFFD4', '#B0E0E6', '#ADD8E6', '#6495ED', '#B0C4DE', '#00BFFF', '#7B68EE', '#87CEFA', '#1E90FF']
plt.pie(temperatures, labels=cities, autopct='%1.1f%%', colors= colors)
plt.title('City Temperatures - Pie Chart')
plt.show()
```
